```python
import json

# 1. List your files in the order you want them merged
files = ["1_eda_regression.ipynb", "2_RF_boosting.ipynb", "3_neural_networks_text_mining.ipynb"]
output_file = "Final_project_code.ipynb"

# 2. Open the first file to use as the "Base"
# (We keep its metadata, kernel info, etc.)
with open(files[0], 'r', encoding='utf-8') as f:
    merged_notebook = json.load(f)

# 3. Loop through the remaining files and just add their cells
for filename in files[1:]:
    with open(filename, 'r', encoding='utf-8') as f:
        notebook = json.load(f)
        # Extend the 'cells' list of the base notebook
        merged_notebook['cells'].extend(notebook['cells'])

# 4. Save the result
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(merged_notebook, f, indent=4)

print(f"Success! Merged into {output_file}")
```
